```python
# ==========================================
# ðŸŽ“ Orthogonal Visual Analytics (Show Mode)
# A Didactic Mini-Project with NumPy, Pandas, Matplotlib
# ==========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Consistent sizing for cleaner viewing (optional)
plt.rcParams["figure.figsize"] = (8, 5)

# -------------------------------
# Step 1: Create Sample Dataset
# -------------------------------
np.random.seed(42)
data = {
    "Category": ["A", "B", "C", "D", "E"],
    "Values1": np.random.randint(10, 100, 5),
    "Values2": np.random.randint(20, 120, 5),
    "Values3": np.random.randint(5, 80, 5)
}
df = pd.DataFrame(data)
```

```python
# -------------------------------
# Step 2: Vertical Bar Chart
# -------------------------------
plt.figure()
plt.bar(df["Category"], df["Values1"])
plt.title("Vertical Bar Chart - Values1")
plt.xlabel("Category")
plt.ylabel("Values1")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 3: Horizontal Bar Chart
# -------------------------------
plt.figure()
plt.barh(df["Category"], df["Values2"])
plt.title("Horizontal Bar Chart - Values2")
plt.xlabel("Values2")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 4: Grouped Bar Chart
# -------------------------------
x = np.arange(len(df["Category"]))
width = 0.3
plt.figure()
plt.bar(x - width/2, df["Values1"], width, label="Values1")
plt.bar(x + width/2, df["Values2"], width, label="Values2")
plt.xticks(x, df["Category"])
plt.title("Grouped Vertical Bar Chart")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 5: Stacked Bar Chart
# -------------------------------
plt.figure()
plt.bar(df["Category"], df["Values1"], label="Values1")
plt.bar(df["Category"], df["Values2"], bottom=df["Values1"], label="Values2")
plt.title("Stacked Vertical Bar Chart")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 6: Horizontal Stacked Bar Chart
# -------------------------------
plt.figure()
plt.barh(df["Category"], df["Values1"], label="Values1")
plt.barh(df["Category"], df["Values2"], left=df["Values1"], label="Values2")
plt.title("Stacked Horizontal Bar Chart")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 7: Line Chart
# -------------------------------
plt.figure()
plt.plot(df["Category"], df["Values3"], marker="o")
plt.title("Line Chart - Values3")
plt.xlabel("Category")
plt.ylabel("Values3")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 8: Multiple Line Chart
# -------------------------------
plt.figure()
plt.plot(df["Category"], df["Values1"], marker="o", label="Values1")
plt.plot(df["Category"], df["Values2"], marker="s", label="Values2")
plt.plot(df["Category"], df["Values3"], marker="^", label="Values3")
plt.title("Multiple Line Chart")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 9: Scatter Plot
# -------------------------------
plt.figure()
plt.scatter(df["Values1"], df["Values2"], marker="x")
plt.title("Scatter Plot - Values1 vs Values2")
plt.xlabel("Values1")
plt.ylabel("Values2")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Step 10: Histogram
# -------------------------------
plt.figure()
plt.hist(df["Values1"], bins=5, alpha=0.7)
plt.title("Histogram - Values1")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
